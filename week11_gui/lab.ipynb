## Section 2: Core Concepts (10 minutes)

### 2.1 What is Recursion?

A **recursive function** is a function that calls itself to solve a smaller version of the same problem.

Every recursive function has two parts:

- **Base case** — the simplest input, where the function stops calling itself.
- **Recursive case** — the function calls itself with a smaller input.

Think of it like opening nested folders: you keep opening one inside another until you reach an empty folder (the base case).

#### Tiny Example: Countdown

```python
# Recursive countdown
def countdown(n):
    if n <= 0:                # base case
        print("Done!")
    else:
        print(n)
        countdown(n - 1)      # recursive case

countdown(3)
```

### 2.2 What is a GUI?

A **GUI** (Graphical User Interface) lets users interact through windows, buttons, and text boxes instead of typing commands.

Python's built-in `tkinter` library lets us create GUIs quickly. The main building blocks are:

- `Tk()` — creates the main window
- `Label` — shows text
- `Entry` — input box for the user
- `Button` — runs a function when clicked

#### Tiny Example: A Hello Window

> **Note:** When you run a `tkinter` cell, a new window will open. **Close that window** before running the next cell.

```python
import tkinter as tk

window = tk.Tk()
window.title("My First GUI")
window.geometry("300x100")

label = tk.Label(window, text="Hello, Cybersecurity Student!")
label.pack(pady=20)

window.mainloop()
```

### 2.3 Connecting a Button to a Function

A button can run a Python function when clicked. We pass the function name to `command=`.

> Remember to close the previous window before running this.

```python
import tkinter as tk

def show_message():
    label.config(text="Button was clicked!")

window = tk.Tk()
window.title("Button Demo")
window.geometry("300x120")

label = tk.Label(window, text="Click the button below")
label.pack(pady=10)

button = tk.Button(window, text="Click Me", command=show_message)
button.pack()

window.mainloop()
```

### Task 1 — Recursive Sum

Write a recursive function `sum_to(n)` that returns the sum of all integers from `1` to `n`.

- Example: `sum_to(4)` should return `10` (because 1 + 2 + 3 + 4 = 10).
- Identify your base case first.
- Do **not** use a loop or the built-in `sum()` function.

Test your function with `sum_to(4)` and `sum_to(10)`.

```python
# Task 1: Write your recursive sum_to(n) function here

def sum_to(n):
    if n <= 1:
      return 1

    else:
      return n + sum_to(n -1)


# Test your function
print(sum_to(4))
print(sum_to(10))
```

### Task 2 — Count Characters in a Password

Write a recursive function `count_chars(text)` that returns the number of characters in a string **without** using `len()`.

- Example: `count_chars("admin")` should return `5`.
- Hint: think about what the "smaller problem" looks like when you remove one character.

This kind of length check is used in password policy validation.

```python
# Task 2: Write your recursive count_chars function here

def count_chars(text):
    if text == "":
      return 0

    else:
      return 1 + count_chars(text[1:])

# Test your function
print(count_chars("admin"))
print(count_chars("P@ssw0rd!"))
```

### Task 3 — Reverse a String Recursively

Write a recursive function `reverse_str(text)` that returns the reversed version of a string.

- Example: `reverse_str("login")` should return `"nigol"`.
- Do **not** use slicing tricks like `text[::-1]`.

Reversing strings is a common step in simple obfuscation and CTF challenges.

```python
# Task 3: Write your recursive reverse_str function here

def reverse_str(text):
    if len(text) <= 1:
      return text

    return reverse_str(text[1:]) + text[0]

# Test your function
print(reverse_str("login"))
print(reverse_str("cybersecurity"))
```
