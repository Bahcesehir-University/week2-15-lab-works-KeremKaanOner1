## Task 5 – Introduction to Sets

Create a list containing the following IP addresses:

192.168.1.1  
192.168.1.2  
192.168.1.1  
10.0.0.5  
10.0.0.5  

Convert this list into a **set**.

Print the set.

```python
ip_addresses = ["192.168.1.1", "192.168.1.2", "192.168.1.1", "10.0.0.5", "10.0.0.5"]
print(ip_addresses)
set_ip = set(ip_addresses)
print(set_ip)
```

## Task 6 – Unique Values

Using the set created in the previous task:

```python
for ip in set_ip:
  print(ip)
print(f"Number of Unique IP Adresses: {len(set_ip)}")
```

## Task 7 – User Input with Lists

Ask the user to enter **five IP addresses**.

Store them in a list.

Then:

- Convert the list into a set
- Print the unique IP addresses
- Print the total number of unique IP addresses

```python
ip_list = []

for i in range(5):
  ip = input("Enter IP address: ")
  ip_list.append(ip)

unique_ips = set(ip_list)

print("unique ips")

for ip in unique_ips:
  print(ip)

print("Total Unique IP: ", len(unique_ips))
```
